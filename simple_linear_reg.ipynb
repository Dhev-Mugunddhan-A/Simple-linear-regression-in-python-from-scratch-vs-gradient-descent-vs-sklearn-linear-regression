```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
X = np.random.randint(5,1000,(1000))
y = np.random.randint(100,2000,1000)
```

```python
plt.scatter(X,y)
```

```python
#min max scaler
X = (X-min(X))/(max(X)-min(X))
y = (y - min(y))/(max(y) - min(y))
```

```python
X_train = X[:800]
y_train= y[:800]
x_test = X[800:]
y_test = y[800:]
n = 800
```

```python
m=(n*sum(X_train*y_train) - (sum(X_train)*sum(y_train)))/(n*(sum(X_train**2))-(sum(X_train)**2))
```

```python
c=(sum(y_train)/n) - (m*(sum(X_train)/n))
```

```python
y_pred = m*x_test + c
```

```python
LSE=(sum((y_pred-y_test)**2))
print(LSE)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,c='red')
plt.show()
```

```python
weights = [1,1]
```

```python
learning_rate = 0.0001
iter_ = 1300
```

```python
Least_square = np.arange(iter_)
```

```python
#X_train
```

```python
for i in range(iter_):
    pred_y = weights[0] + weights[1]*X_train
    error = (y_train - pred_y)
    weights[0] = weights[0] + learning_rate*(2)*np.sum(error)
    weights[1] = weights[1] + learning_rate*(2)*np.sum(error*X_train)
    Least_square[i]=(sum(error**2))
    
```

```python
Least_square[iter_-1]
```

```python
plt.plot(Least_square)
```

```python
plt.plot(x_test,(weights[0] + weights[1]*x_test),c='red')
plt.scatter(x_test,y_test)
```

```python
y_pred_sk = regr.predict(x_test.reshape(-1,1))
```
